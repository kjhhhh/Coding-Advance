```python
import collections
class Solution:
    def updateBoard(self, board, click):
        def BFS(start):
            queue = collections.deque()
            queue.append(start)
            while queue:
                nx, ny = queue.popleft()
                print("当前遍历位置为：(%d,%d) " % (nx,ny))
                lei_count = int(0)
                # 遍历周围八个点 若周围都是空白  下、上、右、左、右下、左下、左上、右上
                for x, y in [(nx-1, ny),(nx-1, ny+1),(nx, ny+1),(nx+1,ny+1),(nx+1,ny),\
                    (nx+1, ny-1),(nx, ny-1),(nx-1, ny-1)]:
                    if 0 <= x < len(board) and 0 <= y < len(board[0]) \
                        and board[x][y] == "M":#周围有雷 标记雷+1
                        # board[x][y] = 'X'
                        lei_count += 1
                if lei_count == 0:#遍历完周围八个点 如果没雷 则递归遍历
                    board[nx][ny] = "B"#则当前点为空白点
                    for x, y in [(nx-1, ny),(nx-1, ny+1),(nx, ny+1),(nx+1,ny+1),(nx+1,ny),\
                    (nx+1, ny-1),(nx, ny-1),(nx-1, ny-1)]:#递归遍历周围八个点
                        if 0 <= x < len(board) and 0 <= y < len(board[0]) \
                            and board[x][y] == "E" :#空白点 则点开 并且加入递归遍历队列
                            board[x][y] = "B" # 要被点开，修改成"B"
                            queue.append((x, y))
                else:
                    board[nx][ny] = (lei_count)
                    # board[nx][ny] = str(lei_count)
                print(queue)
                for i in range(len(board)):
                    print(board[i])
                print('-------------------------')
                    
            
        
            # print(board)
            return board
        # 如果踩雷
        if board[click[0]][click[1]] == "M":
            board[click[0]][click[1]] = "X"
            return board
        return BFS(tuple(click))
```

```python
board = [
        ["E","E","E","E","E"],
        ["E","E","M","E","E"],
        ["E","E","E","E","E"],
        ["E","E","E","E","E"]
    ]
click = [3,0]
Solution().updateBoard(board,click)
```
