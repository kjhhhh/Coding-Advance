```python
list = [3,4,5,1,2,3,11,10,9,6,7,0]
# 1、比较相邻的元素。如果第一个比第二个大，就交换它们两个；
# 2、对每一对相邻元素作同样的工作，从开始第一对到结尾的最后一对，这样在最后的元素应该会是最大的数；
# 3、针对所有的元素重复以上的步骤，除了最后一个；
# 4、重复步骤1~3，直到排序完成。
# 冒泡排序 1、将最大的沉到底
print('冒泡排序 1、将最大的沉到底')
for i in range(len(list)-1,0,-1):
    isSwap = False
    # print(i)
    for j in range(i):
        # print(j)
        if list[j] > list[j+1]:
            list[j],list[j+1] = list[j+1],list[j]
            isSwap = True
    if isSwap == False:
        break;
    print(list)
    print('-------------')
# 冒泡排序 2、将最小的冒泡到顶
list = [3,4,5,1,2,3,11,10,9,6,7,0]
print('冒泡排序 2、将最小的冒泡到顶')
for i in range(len(list)):
    isSwap = False
    # print(i)
    for j in range(len(list)-1,i,-1):
        # print(j)
        if list[j] < list[j-1]:
            list[j],list[j-1] = list[j-1],list[j]
            isSwap = True
    if isSwap == False:
        break;
    print(list)
    print('-------------')
```

```python
list = [3,4,5,1,2,3,11,10,9,6,7,0]
# 1、在未排序序列中找到最小（大）元素，存放到排序序列的起始位置
# 2、从剩余未排序元素中继续寻找最小（大）元素，然后放到已排序序列的末尾。
# 3、复第二步，直到所有元素均排序完毕。
# 选择排序 每次选择出最小值
print('选择排序 1、每次选择出最小值')
for i in range(len(list)):
    Min = i
    for j in range(i+1,len(list)):
        if list[Min] > list[j]:
            Min = j
    if Min!=i:
        list[Min],list[i] = list[i],list[Min]
    print(list)
    print('---------')
    
list = [3,4,5,1,2,3,11,10,9,6,7,0]
# 选择排序 每次选择出最大值
print('选择排序 2、每次选择出最大值')
for i in range(len(list)-1,0,-1):
    Max = i
    for j in range(i-1,0,-1):
        if list[Max] < list[j]:
            Max = j
    if Max!=i:
        list[Max],list[i] = list[i],list[Max]
    print(list)
    print('---------')
```

```python
# 1、把待排序的数组分成已排序和未排序两部分，初始的时候把第一个元素认为是已排好序的。
# 2、从第二个元素开始，在已排好序的子数组中寻找到该元素合适的位置并插入该位置。
# 3、重复上述过程直到最后一个元素被插入有序子数组中。
print('插入排序')
list = [3,4,5,1,2,3,11,10,9,6,7,0]
for i in range(1,len(list)): # i 为未排序的数据 i之前为已排序的数据 即[已排序数据|未排序数据]
    value = list[i]
    index = i
    while index > 0 and list[index-1] > value:#如果当前未排序数据 比已排序数据要小 则将对比的数据后移 直到找到插入位置
        list[index] = list[index-1]#后移
        index-=1
    list[index] = value #将当前遍历数据插入所属位置
    print(list)
    print('---------')
```

```python
# 将序列每相邻两个数字进行归并操作，形成ceil(n/2)个序列，排序后每个序列包含两/一个元素
# 若此时序列数不是1个则将上述序列再次归并，形成ceil(n/4)个序列，每个序列包含四/三个元素
# 重复步骤2，直到所有元素排序完毕，即序列数为1
def MergeSort(list,temp,left,right):
    if left < right:
        mid = (left+right)>>1
        MergeSort(list,temp,left,mid)
        MergeSort(list,temp,mid+1,right)
        temp = Merge(list,temp,left,mid,right)
        
def Merge(list,temp,left,mid,right):
    i,j = left,mid+1
    k = 0
    while i <= mid and j <= right:
        if list[i] <= list[j]:
            temp[k] = list[i]
            k+=1
            i+=1
        else:
            temp[k] = list[j]
            k+=1
            j+=1
        print(temp)
        print('----------')
    while i <= mid:
        temp[k] = list[i]
        k+=1
        i+=1
        print(temp)
        print('----------')
    while j <= right:
        temp[k] = list[j]
        k+=1
        j+=1
        print(temp)
        print('----------')
    for i in range(k):
        list[left+i] = temp[i]
```

```python
#1、从数列中挑出一个元素，称为"基准"（pivot），
#2、重新排序数列，所有比基准值小的元素摆放在基准前面，所有比基准值大的元素摆在基准后面（相同的数可以到任何一边）。在这个分区结束之后，该基准就处于数列的中间位置。这个称为分区（partition）操作。
#3、递归地（recursively）把小于基准值元素的子数列和大于基准值元素的子数列排序。
# 快速排序 
import random
class Solution:
    def sortArray(self, nums):

        def partition(arr, low, high):
            pivot = arr[low]                                        # 选取最左边为pivot

            left, right = low, high     # 双指针
            while left < right:
                
                while left<right and arr[right] >= pivot:          # 找到右边第一个<pivot的元素
                    right -= 1
                arr[left] = arr[right]                             # 并将其移动到left处
                
                while left<right and arr[left] <= pivot:           # 找到左边第一个>pivot的元素
                    left += 1
                arr[right] = arr[left]                             # 并将其移动到right处
            arr[left] = pivot           # pivot放置到中间left=right处
            print(arr)
            print(left)
            print('-----------------------------------')
            return left
        
        def randomPartition(arr, low, high):
            pivot_idx = random.randint(low, high)                   # 随机选择pivot
            arr[low], arr[pivot_idx] = arr[pivot_idx], arr[low]     # pivot放置到最左边
            return partition(arr, low, high)                        # 调用partition函数

        def quickSort(arr, low, high):
            if low >= high:             # 递归结束
                return  
            mid = partition(arr, low, high)       # 以mid为分割点【非随机选择pivot】
            # mid = randomPartition(arr, low, high)   # 以mid为分割点【随机选择pivot】
            quickSort(arr, low, mid-1)              # 递归对mid两侧元素进行排序
            quickSort(arr, mid+1, high)
        

        quickSort(nums, 0, len(nums)-1)             # 调用快排函数对nums进行排序
        return nums


    
```

```python
def heap_sort(array):
    first = len(array) // 2 - 1
    for start in range(first, -1, -1):
        # 从下到上，从右到左对每个非叶节点进行调整，循环构建成大顶堆
        big_heap(array, start, len(array) - 1)
    print('============')
    for end in range(len(array) - 1, 0, -1):#每次取出堆顶元素为已排序数据 并将堆底元素填充到空白堆顶 重新调整堆
        # 交换堆顶和堆尾的数据
        array[0], array[end] = array[end], array[0]
        # 重新调整完全二叉树，构造成大顶堆
        big_heap(array, 0, end - 1)
    return array
 
 
def big_heap(array, start, end):#用于对比根节点与孩子节点的大小 不断调整堆结构
    root = start
    # 左孩子的索引
    child = root * 2 + 1
    while child <= end:
        # 节点有右子节点，并且右子节点的值大于左子节点，则将child变为右子节点的索引
        if child + 1 <= end and array[child] < array[child + 1]:
            child += 1
        if array[root] < array[child]:
            # 交换节点与子节点中较大者的值
            array[root], array[child] = array[child], array[root]
            # 交换值后，如果存在孙节点，则将root设置为子节点，继续与孙节点进行比较
            root = child
            child = root * 2 + 1
        else:
            break
        print(array)
        print('------------')
 
 
# if __name__ == '__main__':
#     array = [10, 17, 50, 7, 30, 24, 27, 45, 15, 5, 36, 21]
#     print(heap_sort(array))
```

```python
# 1、找出待排序数组中的最大值max、最小值min
# 2、我们使用 动态数组ArrayList 作为桶，桶里放的元素也用 ArrayList 存储。桶的数量为(max-min)/arr.length+1
# 3、遍历数组 arr，计算每个元素 arr[i] 放的桶
# 4、每个桶各自排序
# 5、遍历桶数组，把排序好的元素放进输出数组
print('桶排序')
def bucktetSort(numList,bucketSize):
    if len(numList) == 0 or len(numList) == 1:
        return numList
    maxNum = max(numList)
    minNum = min(numList)
    bucketNum = (maxNum - minNum + 1) // bucketSize+1   # 根据桶的数量找到每个桶的范围
    buckets = [[] for i in range(bucketNum)]
    # print(bucketNum)
    # print(buckets)
    for i in range(len(numList)):     # 将各个数分配到各个桶
        # print((numList[i] - minNum) // bucketSize)
        buckets[(numList[i] - minNum) // bucketSize].append(numList[i])
    print(buckets)
    print('---------------')
    for i in range(bucketNum):       # 桶内排序，可以使用各种排序方法
        buckets[i].sort()
    print(buckets)
    print('---------------')
    res = []
    for i in range(len(buckets)):   # 分别将各个桶内的数提出来，压入结果
        for j in range(len(buckets[i])):
            res.append(buckets[i][j])
            print(res)
        print('---------------')
    return res
list = [3,4,5,1,2,3,11,10,9,6,7]
numlist = [11,34,23,56,8,20,66,45,54,87,78]
print(bucktetSort(list,3))  # 利用5个桶
# 输出结果为：[8, 11, 20, 23, 34, 45, 54, 56, 66, 78, 87]
```

```python
# coding=utf-8
def bucket_sort2(array):
    min_num, max_num = min(array), max(array)
    bucket_num = (max_num-min_num)//3 + 1
    buckets = [[] for _ in range(int(bucket_num))]
    # print(bucket_num)
    # print(buckets)
    for num in array:
        buckets[int((num-min_num)//3)].append(num)
    new_array = []
    for i in buckets:
        for j in sorted(i):
            new_array.append(j)
    return new_array
list = [3,4,5,1,2,3,11,10,9,6,7]
numlist = [11,34,23,56,8,20,66,45,54,87,78]
print(bucket_sort2(list))  # 利用5个桶
```
